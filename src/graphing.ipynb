```python
import gymnasium as gym
import mujoco

env = gym.make("Humanoid-v5", xml_file="../models/custom_humanoid_2xSluggish.xml")

model = env.unwrapped.model
num_bodies = model.nbody

# Exclude "world" (index 0)
body_names = [mujoco.mj_id2name(model, mujoco.mjtObj.mjOBJ_BODY, i) for i in range(1, num_bodies)]

print(body_names)
```
